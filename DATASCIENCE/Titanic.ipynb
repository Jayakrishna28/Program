```python
d=pd.get_dummies(df1['Sex'])
d=d.astype(int)
d.head()
```

```python
df1=pd.concat([df1,d],axis='columns')
df1.head()
```

```python
df1=df1.drop(['Sex','male'],axis='columns')
df1.head()
```

```python
X=df1.drop(['Survived'],axis='columns')
X.head()
```

```python
y=df1['Survived']
y.head()
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```
