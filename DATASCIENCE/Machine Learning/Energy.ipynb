```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("C:/Users/LENOVO/Downloads/train_energy_data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df1=pd.get_dummies(df['Building Type'])
df1=df1.astype(int)
df1.head()
```

```python
df2=pd.get_dummies(df['Day of Week'])
df2=df2.astype(int)
df2.head()
```

```python
df3=pd.concat([df,df1,df2],axis='columns')
df3.head()
```

```python
df4=df3.drop(['Building Type','Day of Week','Residential','Weekday'],axis='columns')
df4.head()
```

```python
X=df4.drop(['Energy Consumption'],axis='columns')
X.head()
```

```python
y=df4['Energy Consumption']
y.head()
```

```python
m=linear_model.LinearRegression()
m.fit(X,y)
```

```python
m.predict([[7063,76,10,29.84,0,0,0]])
```
